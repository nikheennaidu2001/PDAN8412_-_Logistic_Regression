### Building the Logistic Regression Model

In this step we'll build a Logistic Regression Model from scratch using the Sigmoid and Gradient Descent formulas

```python
#Code Borrowed from:
#AssemblyAI
#2022
#How to implement Logistic Regression from scratch with Python
#Software: Python 
#( Youtube, link: https://youtu.be/YYEJ_GUguHw?si=skj3DkhFByAf8Ifj )


import numpy as np # USED FOR STANDARD LINEAR ALGEBRA

# 'sigmoid' IS A CLASS OF FUNCTIONS, FROM LINEAR REGRESSION, THAT INTRODUCES NON-LINEARITY INTO THE MODEL 
# IT'S MAIN FUCTION IS TO MAP THE DATA INTO A PROBABILITY OF 1 AND 0

def sigmoid(x):                 # TAKEN FROM THE FORMULA  fx =   1
    return 1/(1+np.exp(-x))     #                            _________
                                #                             1 + e^-x

# CREATE A CLASS FOR THE MODEL THAT'LL BE CALLED LATER
class LogisticRegression() :

# 'lr' LEARNING RATE DETERMINES HOW RADIDLY WE NEED TO UPDATE THE PARAMETERS
# WE SET lr=0.001 BECAUSE IT INCREASES THE TIME SPENT ON TRAINING THE MODEL, AND THE VARIANCE IN THE TRAINING TIME

# 'n_iters' IS THE NUMBER OF PARAMETER SETTINGS THAT'S SAMPLED
    def __init__(self, lr=0.001, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None  # 'weights' IS A PARAMETER FOR WEIGHING THE POSSIBILITY OF EACH VALUE
        self.bias = None    # 'bias' IS THE DIFFERENCE BETWEEN PREDICTED AND ACTUAL VALUES

    # METHOD TO FIT THE MODEL
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0


        for _ in range(self.n_iters):
            linear_pred = np.dot(X, self.weights) + self.bias # WE CREATE A VARIABLE FOR LINEAR PREDICTIONS
            predictions = sigmoid(linear_pred) 
            # GRADIENTS
            dw = (1/n_samples) * np.dot(X.T, (predictions - y)) # FORMULA FOR GRADIENT WEIGHT
            db = (1/n_samples) * np.sum(predictions-y) # FORMULA FOR GRADIENT BIAS

            self.weights = self.weights - self.lr*dw
            self.bias = self.bias - self.lr*db

    # METHOD FOR PREDICTIONS
    def predict(self, X):
        linear_pred = np.dot(X, self.weights) + self.bias
        y_pred = sigmoid(linear_pred)
        class_pred = [0 if y<=0.5 else 1 for y in y_pred] 
        return class_pred
```
